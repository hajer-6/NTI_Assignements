```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, regularizers
np.random.seed(0); tf.random.set_seed(0)
print("TensorFlow:", tf.__version__)

# ---- PROVIDED: load MNIST, keep only 1000 training images (so it overfits fast) ----
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
x_train = x_train.reshape(-1, 784).astype("float32") / 255.0
x_test  = x_test.reshape(-1, 784).astype("float32") / 255.0
x_small, y_small = x_train[:1000], y_train[:1000]
print("small training set:", x_small.shape, "| test set:", x_test.shape)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(x_small.shape[1],)),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history = model.fit(
    x_small,
    y_small,
    epochs=100,
    validation_data=(x_test, y_test),
)

train_acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(x_small.shape[1],)),
    tf.keras.layers.Dense(512, activation='relu',kernel_regularizer=regularizers.l2(1e-4)),
    tf.keras.layers.Dense(512, activation='relu',kernel_regularizer=regularizers.l2(1e-4)),
    tf.keras.layers.Dense(512, activation='relu',kernel_regularizer=regularizers.l2(1e-4)),
    tf.keras.layers.Dense(10, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history = model.fit(
    x_small,
    y_small,
    epochs=100,
    validation_data=(x_test, y_test),
)

train_acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
```

```python
early_stopping = keras.callbacks.EarlyStopping(
    monitor="val_loss", patience=5, restore_best_weights=True
)

model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(x_small.shape[1],)),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history = model.fit(
    x_small,
    y_small,
    epochs=100,
    validation_data=(x_test, y_test),
    callbacks=[early_stopping]
)

train_acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
```
