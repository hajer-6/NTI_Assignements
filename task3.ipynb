# 1- VGG_16

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.utils import to_categorical
```

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
y_train, y_test = to_categorical(y_train, 10), to_categorical(y_test, 10)
```

```python
def VGG_16(input_shape=(32,32,3),num_classes=10):
  model=Sequential()
  #block-1
  model.add(Conv2D(64, (3, 3), activation='relu', padding='same', input_shape=input_shape))
  model.add(Conv2D(64, (3, 3), activation='relu', padding='same'))
  model.add(MaxPooling2D())
  #block-2
  model.add(Conv2D(128, (3, 3), activation='relu', padding='same'))
  model.add(Conv2D(128, (3, 3), activation='relu', padding='same'))
  model.add(MaxPooling2D())
  #block-3
  model.add(Conv2D(256,(3,3),activation='relu',padding='same'))
  model.add(Conv2D(256,(3,3),activation='relu',padding='same'))
  model.add(Conv2D(256,(3,3),activation='relu',padding='same'))
  model.add(MaxPooling2D())
  #block-4
  model.add(Conv2D(512,(3,3),activation='relu',padding='same'))
  model.add(Conv2D(512,(3,3),activation='relu',padding='same'))
  model.add(Conv2D(512,(3,3),activation='relu',padding='same'))
  model.add(MaxPooling2D())
  #block-5
  model.add(Conv2D(512,(3,3),activation='relu',padding='same'))
  model.add(Conv2D(512,(3,3),activation='relu',padding='same'))
  model.add(Conv2D(512,(3,3),activation='relu',padding='same'))
  model.add(MaxPooling2D())
  #fully connected
  model.add(Flatten())
  model.add(Dense(4096,activation='relu'))
  model.add(Dense(4096,activation='relu'))
  model.add(Dense(num_classes,activation='softmax'))
  return model
```

```python
model16=VGG_16()
model16.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
history=model16.fit(x_train,y_train,batch_size=32,epochs=1,validation_data=(x_test,y_test))
predictions = model16.predict(x_test)
```

# 2- VGG_19

```python
def VGG_19(input_shape=(32,32,3),num_classes=10):
  model=Sequential()
  #block-1
  model.add(Conv2D(64, (3, 3), activation='relu', padding='same', input_shape=input_shape))
  model.add(Conv2D(64, (3, 3), activation='relu', padding='same'))
  model.add(MaxPooling2D())
  #block-2
  model.add(Conv2D(128, (3, 3), activation='relu', padding='same'))
  model.add(Conv2D(128, (3, 3), activation='relu', padding='same'))
  model.add(MaxPooling2D())
  #block-3
  model.add(Conv2D(256,(3,3),activation='relu',padding='same'))
  model.add(Conv2D(256,(3,3),activation='relu',padding='same'))
  model.add(Conv2D(256,(3,3),activation='relu',padding='same'))
  model.add(Conv2D(256,(3,3),activation='relu',padding='same'))
  model.add(MaxPooling2D())
  #block-4
  model.add(Conv2D(512,(3,3),activation='relu',padding='same'))
  model.add(Conv2D(512,(3,3),activation='relu',padding='same'))
  model.add(Conv2D(512,(3,3),activation='relu',padding='same'))
  model.add(Conv2D(512,(3,3),activation='relu',padding='same'))
  model.add(MaxPooling2D())
  #block-5
  model.add(Conv2D(512,(3,3),activation='relu',padding='same'))
  model.add(Conv2D(512,(3,3),activation='relu',padding='same'))
  model.add(Conv2D(512,(3,3),activation='relu',padding='same'))
  model.add(Conv2D(512,(3,3),activation='relu',padding='same'))
  model.add(MaxPooling2D())
  #fully connected
  model.add(Flatten())
  model.add(Dense(4096,activation='relu'))
  model.add(Dense(4096,activation='relu'))
  model.add(Dense(num_classes,activation='softmax'))
  return model
```

```python
model19=VGG_19()
model19.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
history=model19.fit(x_train,y_train,batch_size=32,epochs=1,validation_data=(x_test,y_test))
predictions=model19.predict(x_test)
```

```python
from re import M
def Alex_net(input_shape=(227,227,3),num_classes=10):
  model=Sequential()
  #layer1
  model.add(Conv2D(96,(11,11),strides=(4,4),activation='relu',padding='same'))
  model.add(MaxPooling2D())
  #layer2
  model.add(Conv2D(256,(5,5),activation='relu',padding='same'))
  model.add(MaxPooling2D())
  #layer3
  model.add(Conv2D(384,(3,3),activation='relu',padding='same'))
  model.add(MaxPooling2D())
  #layer4
  model.add(Conv2D(384,(3,3),activation='relu',padding='same'))
  model.add(MaxPooling2D())
  #layer5
  model.add(Conv2D(256,(3,3),activation='relu',padding='same'))
  model.add(MaxPooling2D())
  #fully connected
  model.add(Flatten())
  model.add(Dense(4096,activation='relu'))
  model.add(Dense(4096,activation='relu'))
  model.add(Dense(num_classes,activation='softmax'))
  return model
```

```python
modelA=Alex_net()
modelA.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
history=modelA.fit(x_train,y_train,batch_size=32,epochs=1,validation_data=(x_test,y_test))
predictions=modelA.predict(x_test)
```
